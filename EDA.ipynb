```python
# Imports
import glob
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
import random
```

# Plot the dimensions

```python
def get_dims(file):
    """Returns dimensions for an RBG image"""
    im = Image.open(file)
    arr = np.array(im)
    h,w,d = arr.shape
    return h,w

numOfDataEnteries = 12499
print(random.randint(0,12499))

catDimsx = []
catDimsy = []

for i in range(125):
    file = 'PetImages/Cat/{}.jpg'.format(random.randint(0,numOfDataEnteries))
    picDims = get_dims(file)
    catDimsx.append(picDims[1])
    catDimsy.append(picDims[0])

plt.scatter(catDimsx,catDimsy)
plt.title("Cat Image Dimensions")
plt.xlabel("Width")
plt.ylabel("Height")
plt.show()
```

```python
dogDimsx = []
dogDimsy = []

for i in range(125):
    file = 'PetImages/Dog/{}.jpg'.format(random.randint(0,numOfDataEnteries))
    picDims = get_dims(file)
    dogDimsx.append(picDims[1])
    dogDimsy.append(picDims[0])

plt.scatter(dogDimsx,dogDimsy)
plt.title("Dog Image Dimensions")
plt.xlabel("Width")
plt.ylabel("Height")
plt.show()
```

```python
numRows = 2
numColumns = 4
totalImageCount = numRows * numColumns

images = []
for i in range(0,totalImageCount):
    file = 'PetImages/Dog/{}.jpg'.format(random.randint(0,numOfDataEnteries))
    image = Image.open(file)
    images.append(image)

def display_multiple_img(images, rows, cols):
    figure, ax = plt.subplots(nrows=rows,ncols=cols )
    for ind,title in enumerate(images):
        ax.ravel()[ind].imshow(images[ind])
        #ax.ravel()[ind].set_title(title)
        #ax.ravel()[ind].set_axis_off()
    plt.tight_layout()
    #plt.figure(figsize=(8,6), dpi=80)
    plt.rcParams["figure.figsize"] = (6,100)
    plt.show()

display_multiple_img(images, numRows, numColumns)
```
