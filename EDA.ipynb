# Import useful libraries

```python
import glob
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from PIL import Image
import random
```

# Establsh 0.008 of dataset as my working sample size randomy selected for each categorization

```python
numOfDataEntries = 12500
percentOfData = 0.008
catImageSampleSet = []
dogImageSampleSet = []

for i in range(0,int((numOfDataEntries*percentOfData*.5))):
    file = 'PetImages/Dog/{}.jpg'.format(random.randint(0,numOfDataEntries))
    image = Image.open(file)
    dogImageSampleSet.append(image)

    file = 'PetImages/Cat/{}.jpg'.format(random.randint(0,numOfDataEntries))
    image = Image.open(file)
    catImageSampleSet.append(image)

random.shuffle(catImageSampleSet)
random.shuffle(dogImageSampleSet)
totalSampleSet = catImageSampleSet + dogImageSampleSet
```

# Get a feel for the datasets by viewing rows of images.

```python
# Creat viewing sample set
numOfPairsOfRows = 2
viewSampleSet = catImageSampleSet[0:(8*numOfPairsOfRows)] + dogImageSampleSet[0:(8*numOfPairsOfRows)]

# table setup, then show
def display_multiple_img(images, rows, cols):
    figure, ax = plt.subplots(nrows=rows,ncols=cols)
    for ind,title in enumerate(images):
        ax.ravel()[ind].imshow(images[ind])
        #ax.ravel()[ind].imshow(images[title])
        #ax.ravel()[ind].set_title(title)
        ax.ravel()[ind].set_axis_off()
    plt.tight_layout()
    plt.gcf().set_size_inches(20, 16)
    plt.show()

display_multiple_img(viewSampleSet, numOfPairsOfRows*2, 8)
```

# Plot the dimensions of sample data set by subdivisions to understand overall size

```python
"""Returns dimensions for image"""
def get_dims(image):
    arr = np.array(image)
    h = arr.shape[0]
    w = arr.shape[1]
    return h,w

catDimsWidith = []
catDimsHeight = []

for i in range(len(catImageSampleSet)):
    imageDims = get_dims(catImageSampleSet[i])
    catDimsWidith.append(imageDims[0])
    catDimsHeight.append(imageDims[1])

plt.scatter(catDimsWidith,catDimsHeight)
plt.title("Cat Image Dimensions")
plt.xlabel("Width")
plt.ylabel("Height")
plt.show()
```

```python

dogDimsWidith = []
dogDimsHeight = []

for i in range(len(dogImageSampleSet)):
    imageDims = get_dims(dogImageSampleSet[i])
    catDimsWidith.append(imageDims[0])
    catDimsHeight.append(imageDims[1])

plt.scatter(catDimsWidith,catDimsHeight)
plt.title("Cat Image Dimensions")
plt.xlabel("Width")
plt.ylabel("Height")
plt.show()
```
